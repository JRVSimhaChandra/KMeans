```python
# 3-sigma rule
low_limit=data['BloodPressure'].mean()-3*data['BloodPressure'].std()
upper_limit=data['BloodPressure'].mean()+3*data['BloodPressure'].std()
print("lower limit",low_limit)
print("upper limit",upper_limit)
```

```python
# FInd values less than lower limit
data.loc[data['Insulin']<low_limit]
```

```python
# Find values greater than upper limit
data.loc[data['BloodPressure']>upper_limit]
```

```python
# Replace outliers with mean
data.loc[data['BloodPressure']>upper_limit,'BloodPressure']=data['BloodPressure'].mean()
```

```python
sns.boxplot(data=data,x='BloodPressure')
```

```python
# Apply Scaling
from sklearn.preprocessing import StandardScaler
scaling=StandardScaler()
data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']]=scaling.fit_transform(data[['Pregnancies', 'Glucose', 'BloodPressure', 'SkinThickness', 'Insulin',
       'BMI', 'DiabetesPedigreeFunction', 'Age']])
```

```python
# Remove outcome from data
x=data.drop('Outcome',axis=1)
```

# Apply Kmean model

```python
# Import kmean model
from sklearn.cluster import KMeans
# initialize the model
model=KMeans(n_clusters=6,random_state=10)
# train a model 
model.fit(x)
```

```python
# centroids-cluster centers
# no of clusters is sames as number of centroids
model.cluster_centers_
```

```python
model.labels_
```

```python
# Visualize the clusters created by model
color_schema=np.array(["red","green","black","orange","blue","pink"])
plt.scatter(data.Insulin,data.Glucose,color=color_schema[model.labels_])
```

```python
plt.figure(figsize=(3,3))
color_schema=np.array(["red","green","black","orange","blue","pink"])
plt.scatter(data.Insulin,data.Glucose,color=color_schema[model.labels_])
```

```python
plt.figure(figsize=(3,3))
color_schema=np.array(["red","green","black","orange","blue","pink"])
plt.scatter(x.Insulin,x.Glucose,color=color_schema[model.labels_])
```

```python
data['labels']=pd.DataFrame(model.labels_)
```
